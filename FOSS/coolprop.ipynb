O QUE É COOLPROP?


CoolProp é uma biblioteca C++ que implementa:


*   Equações de estado e propriedades de transporte de fluidos puros e pseudo-puros para 122 componentes
*   Propriedades da mistura usando formulações de energia Helmholtz de alta precisão
*   Correlações de propriedades de fluidos incompressíveis e salmouras
*   Interpolação tabular computacionalmente eficiente
*   Rotinas psicrométricas de maior precisão
*   Rápido IAPWS-IF97 (Formulação Industrial) para Água/Vapor
*   Equações cúbicas de estado (SRK, PR)

## EXEMPLO DE USO DO COOLPROP

DENSIDADE DE UM GÁS

A densidade de um gás, pode ser dada pela seguinte fórmula:

***colocar imagem
{\display d ={ \frac {{Mp} {TR}}}

Onde:

    d é a densidade, em quilogramas por metros cúbicos (kg/m3);

    M é a quantidade de matéria, em moles (mol);

    p é a pressão, em pascais (Pa);

    T é a temperatura, em kelvins (K°);

    R é a constante dos gases reais, equivalente a 8,31 (N/m2).m3/mol.K.

## UTILIZANDO COOLPROP para o cálculo da Densidade

Um código para calcular a densidade D do nitrogênio a uma temperatura T de 298 K e uma pressão P de 101325 Pa é:

densidade = PropsSI('D', 'T', 298.15, 'P', 101325, 'Nitrogen')

EXEMPLO

```python
pip install CoolProp #instalando CoolProp
```

```python
#Importando Sistema Internacional de Unidades do CoolProp
from CoolProp.CoolProp import PropsSI
'''
Cálculo da densidade do nitrogênio
Parâmentros:
P1: 10^5 Pascal
T1: 0°C
'''
P1 = 101325 #Pa
T1 = 298.15 #Kelvin
D1 = PropsSI('D','T',T1,'P',P1,'nitrogen')
print("A densidade é de", int(D1), "kg/m³")
```

```python
'''
Cálculo da densidade da água
Parâmentros:
P1: 10^5 Pascal
T1: 0°C
'''
P1 = 10e5 #Pa
T1 = 0 + 273.15 #°C para K
D1 = PropsSI('D','T',T1,'P',P1,'water')
print("A densidade é de", int(D1), "kg/m³")
```

```python
'''
Cálculo da densidade do Metanol
Parâmentros:
P1: 10^5 Pascal
T1: 40°C
'''
P1 = 10e5 #Pa
T1 = 40 + 273.15 #°C para K
D1 = PropsSI('D','T',T1,'P',P1,'methanol')
print("A densidade é de", int(D1), "kg/m³")
```
