```python
formula = Formula('H2O') # Comandos para o cálculo da massa molar de determinada molécula.
formula.mass  # g/mol
print(f'{formula.mass} g/mol') # Água
```

#**Mendeleev**

*Esta é uma biblioteca Python muito útil no curso de Estrutura da Matéria. Ela permite acessar várias propriedades químicas dos elementos, como:*

 • *Acesso a tabela periódica;*

 • *Ponto de ebulição dos elementos químicos;*

 *Dentre outras funções que serão abordadas a seguir.*

```python
pip install mendeleev # Comando utilizado para instalar a biblioteca Mendeleev para o uso dela no curso de Estrutura da Matéria.
```

```python
from mendeleev.vis import periodic_table # Comando para a importação da tabela periódica.
periodic_table()
```

```python
from mendeleev import Ne # Comando para a amostra do ponto de ebulição de determinado átomo, sendo esse ponto na unidade Kelvin(Sistema Internacional).
Ne.boiling_point  # K 
print(f'{Ne.boiling_point} K')
```

```python
from mendeleev import Al # Comandos para a amostra do ponto de fusão de determinado átomo, sendo esse ponto na unidade Kelvin(Sistema Internacional).
Al.melting_point # K
print(f'{Al.melting_point} K')
```

```python
from mendeleev import element # Comando de importação
```

```python
from mendeleev import Cl # Comandos para encontrar o raio atômico de determinado átomo, sendo esse raio em picômetros.
Cl = element(17)
Cl.atomic_radius  # pm
print(f'{Cl.atomic_radius} pm')
```

```python
from mendeleev import Og # Comandos para a amostra da configuração eletrônica de determinado átomo.
Oganesson = element(118)
Oganesson.econf
```

```python
from mendeleev import Fe  # Comandos para a amostra da configuração eletrônica de determinado íon.
ferro = element(26)
ferro.ec.ionize(3) #Fe³+
```

```python
from mendeleev import C # Comandos para a amostra dos estados de oxidação(Nox) de determinado átomo.
carbono = element('C')
carbono.oxistates
```

#**Chempy**
*ChemPy é uma biblioteca Python que permite resolver inúmeras funções na área da química, porém para o estudo do curso da Estrutura da Matéria, vale ressaltar o uso da função de balanceamento estequiométrico.*

```python
pip install chempy # Comando para a instalação da biblioteca Chempy
```

```python
from chempy import balance_stoichiometry # Comandos para a amostra do balançeamento estequiométrico. 
from chempy import Reaction

reac, prod = balance_stoichiometry({'C6H12O6', 'O2'}, {'CO2', 'H2O'})

reaction = Reaction(reac, prod)
reaction
```

```python
reac, prod = balance_stoichiometry({'Na2CO3', 'HCl'}, {'NaCl', 'H2O', 'CO2'}) # Comandos para a amostra do balançeamento estequiométrico.

reaction = Reaction(reac, prod)
reaction
```

```python
reac, prod = balance_stoichiometry({'CuSO4', 'Fe'}, {'FeSO4', 'Cu'}) # Comandos para a amostra do balançeamento estequiométrico.

reaction = Reaction(reac, prod)
reaction
```

```python
reac, prod = balance_stoichiometry({'BaCl2', 'Al2(SO4)3'}, {'BaSO4', 'AlCl3'}) # Comandos para a amostra do balançeamento estequiométrico.

reaction = Reaction(reac, prod)
reaction
```
