<font size="10px">Linguagens e Comandos</font>

&nbsp;&nbsp;&nbsp;&nbsp;
O MySQL oferece todos os recursos e ferramentas de SGBDs fechados, como Triggers, Store Procedures, Rules, Constraints e outros. A ferramenta também possui sublinguagens com comandos e operações para a aplicação total dos dados:
- 
<font color="#44f">DDL</font>
<font color="#aaa">(Data Definition Language)</font>: 
Linguagem de Definição de Dados, define estruturas do banco (tabelas, visões, índices, sequências, etc);
 - CREATE: Cria uma estrutura;
 - ALTER: Altera uma estrutura;
 - DROP: Exclui uma estrutura.
- 
<font color="#4af">DML</font>
<font color="#aaa">(Data Manipulation Language)</font>: 
Linguagem de Manipulação de Dados, gerencia os dados populados/a popular as estrutras criadas no DDL;
 - INSERT: Insere dados;
 - UPDATE: Altera dados;
 - DELETE: Excluid dados.
- 
<font color="#4ff">DQL</font>
<font color="#aaa">(Data Query Language)</font>: 
Linguagem de Consulta de Dados, comando principal, pelo uso de consulta e retorno de dados do banco;
 - SELECT: Retorna dados, os ordena, agrupa, aplica filtros e mais.
- 
<font color="#4fa">DCL</font>
<font color="#aaa">(Data Control Language)</font>: 
Linguagem de Controle de Dados, controla o acesso aos dados e operações dos usuários;
 - GRANT: Habilita acesso a dados e operações;
 - REVOKE: Remove acesso a dados e operações;
- 
<font color="#4f4">DTL</font>
<font color="#aaa">(Data Transaction Language)</font>: 
Linguagem de Transação de Dados, gerencia as transações (Conjunto de operações e comandos);
 - START TRANSACTION: Inicia a transação;
 - COMMIT: Concretiza a transação no banco;
 - ROLLBACK: Anula a transação.
