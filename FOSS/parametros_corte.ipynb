Este notebook tem como objetivo facilitar o calculo dos parametros de corte para ferramentas rotativas na linha de produção da industria de usinagem.

As variaveis utilizadas nestes calculos serão:

n= Frequência de rotação (n) [rpm]
- É o número de voltas por unidade de tempo que a fresa dá em torno do seu eixo.

Vf= velocidade do avanço em mm/min
- É a velocidade instantânea do ponto selecionado sobre o gume, no movimento de avanço, em relação a peça. No fresamento, o movimento de avanço é provocado pela translação da ferramenta sobre a peça ou vice-versa. A direção da velocidade de avanço é, então, radial ao eixo da ferramenta.

Vc= velocidade de corte em m/min
- É a velocidade instantânea do ponto selecionado sobre o gume, no movimento de corte, em relação a peça. No fresamento, o movimento de corte é proporcionado pela rotação da ferramenta. A velocidade de corte é, então, uma velocidade tangencial. As grandezas relacionadas ao movimento de corte recebem o índice “c”


Z= É o número total de dentes que a fresa contém.


d= É o diâmetro externo da fresa.


fz= avanço por dente em milímetro
- É a distância linear percorrida por um dente da ferramenta no intervalo em que dois dentes consecutivos entram em corte. Também é medido no plano de trabalho.

## Cálculos de rotação de ferramentas para fresamento, utilizando o exemplo acima na condição:
## material=aço inoxidável, inserto= classe F7030, método de usinagem=C,E,F

```python
import numpy as np
```

Fórmula para encontrar RPM da ferramenta dado a velocidade de corte e o diâmetro da ferramenta
$$\text{n}= \frac{1000\times Vc}{d  \times π}   $$ 

```python
Vc=200  #velocidade de corte
d=60  #Diâmetro da ferramenta
n=(1000*Vc)/(d*np.pi)  #fórmula para retornar rotações por minuto
print(f'{n:.3f} RPM')
```

Fórmula para encontrar a velocidade de corte em metros por minuto dado RPM e o diâmetro da ferramenta
$$\text{Vc}= \frac{n\times d\times \ π }{1000} $$ 

```python
n=n  #Rotações por minuto
d=d  #Diâmetro da ferramenta
Vc=(n*d*np.pi)/1000  #fórmula para retornar velocidade de corte
print(f'{Vc:.3f} m/min')
```

## Cálculos de velocidade de avanço

Fórmula para encontrar o avanço de corte em milimetros por minuto, dado Velocidade de corte, o número de dentes da ferramenta e o avanço por dente
$$\text{Vf}= \frac{1000\times Vc\times \ Z\times fz }{d\times π} $$ 

```python
Vc=Vc  #velocidade de corte
Z=4  #número de dentes da ferramenta
Fz=0.1  #avanço por dente em milímetro
Vf=(1000*Vc*Z*Fz)/(d*np.pi)  #fórmula para retornar avanço da ferramenta
print(f'{Vf:.3f} mm/min')
```

Fórmula para encontrar o avanço de corte em milimetros por minuto, dado rotações por minuto, o número de dentes da ferramenta e o avanço por dente
$$\text{Vf}= {n\times \ Z\times fz } $$ 

```python
n=n  #Rotações por minuto
Z=Z  #número de dentes da ferramenta
Fz=Fz  #avanço por dente em milímetro
Vf=n*Z*Fz  #fórmula para retornar avanço da ferramenta
print(f'{Vf:.3f} mm/min')
```
