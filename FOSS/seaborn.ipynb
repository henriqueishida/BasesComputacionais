###  Plotar uma tabela com os dados
Para utilizarmos como exemplo, utilizamos um banco de dados da ANEEL (Agência Nacional de Energia Elétrica) onde é mostrado um levantamento da potência elétrica consumida por, estado brasileiro, em um certo intervalo de tempo.
* Podemos notar que os dados são extremamente atuais, sendo a última postagem no dia 01/12/2022.

```python
df = pd.read_csv('https://raw.githubusercontent.com/Daniellucasgonzaga/Trabalho_Final_BCC19/main/Dados.csv')
df
```

### TIPOS DE GRÁFICOS DO SEABORN:



tipos de graficos.png

### CRIANDO GRÁFICO DE DISPERSÃO (SCATTERPLOT)
DOCUMENTAÇÃO: https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```python
fig = sns.scatterplot(data = df,
                      y = "MdaPotenciaInstaladakW",  
                      x = "SigUF",            
                      hue = "AnoReferencia"
                    )
fig.figure.set_size_inches(20, 10)
```

Como o gráfico acima ficou um pouco confuso, separaremos os dados apenas para o estado de São Paulo e Bahia.

```python
SP = df.query("SigUF == 'SP'")
SP.head()
```

```python
BA = df.query("SigUF == 'BA'")
BA.head()
```

Plotando o grafico de dispersões do estado de São Paulo.

```python
fig1 = sns.scatterplot(data = SP,                      # busca o df
                      x = "MdaPotenciaInstaladakW",    # valor eixo x
                      y = "AnoReferencia",             # valor eixo y
                      hue = "MesReferencia",           # adiciona os meses no grafico (nova variável) 
                      style = "MesReferencia",         # muda estilo dos pontos
                      palette = "deep",                # muda paleta de cores dos pontos
                      size = "MdaPotenciaInstaladakW", # muda tamanho com relação à coluna digitada
                      sizes = (70, 200))               # tamanho do menor, tamanho do maior ponto
fig1.figure.set_size_inches(15, 8)                     # altera tamanho da figura
```

### CRIANDO GRÁFICO DE LINHA (LINEPLOT)
DOCUMENTAÇÃO: https://seaborn.pydata.org/generated/seaborn.lineplot.html
* Utilizando os mesmos dados acima (Potência Instalada em SP), plotamos um gráfico de linhas. 
* Também é plotado o intervalo da potencia instalada para cada mês (hachura azul clara).

```python
line = sns.lineplot(data = SP,                    # busca o df
                    x = "AnoReferencia",          # valor eixo x
                    y = "MdaPotenciaInstaladakW") # valor eixo y
line.figure.set_size_inches(15, 8)                # altera tamanho da figura
```

Outro modelo de plotagem de gráfico de linhas.

```python
line1 = sns.lineplot(data = SP,                                    # busca o df
                     x = "AnoReferencia",                          # valores eixo x
                     y = "MdaPotenciaInstaladakW",                 # valores eixo y
                     hue = "MesReferencia",                        # separa os graficos em meses
                     style = "MesReferencia"                       # altera estilos de linha 
                     ).set_title('Potência instalada em São Paulo')# adiciona título
line1.figure.set_size_inches(15, 8)                                # altera tamanho da figura 
```

### CRIANDO UM HISTOGRAMA (HISTPLOT)
DOCUMENTAÇÃO: https://seaborn.pydata.org/generated/seaborn.histplot.html

Histograma no eixo horizontal

```python
sns.histplot(data = SP,
             y = "MdaPotenciaInstaladakW") 
```

Histograma no eixo vertical

```python
sns.histplot(data = SP,
             x = "MdaPotenciaInstaladakW",
             bins = 30) # altera o número de barras
```

Adicionando curva de densidade

```python
sns.histplot(data = SP,
             x = "MdaPotenciaInstaladakW",
             kde = True) # mostra a curva de densidade
```

Adicionando mais uma variável ao gráfico por sobreposição

```python
sns.histplot(data = SP,
             x = "MdaPotenciaInstaladakW",
             hue = "MesReferencia") # adiciona mais uma variável
```

Para facilitar a visualização dos dados sobrepostos, podemos utilizar uma função "degrau"

```python
sns.histplot(data = SP,
             x = "MdaPotenciaInstaladakW",
             hue = "MesReferencia", # adiciona mais uma variável
             element = "step")      # altera a vizualização para "degrau"
```

Histograma no formato poligonal

```python
sns.histplot(data = SP,
             x = "MdaPotenciaInstaladakW",
             hue = "MesReferencia", # adiciona mais uma variável
             element = "poly")      # altera a vizualização para "poligonal"
```

### CRIANDO GRAFICO KDE (KDEPLOT)
* Plota histograma com apenas uma curva de densidade.

DOCUMENTAÇÃO: https://seaborn.pydata.org/generated/seaborn.kdeplot.html

```python
sns.kdeplot(data = SP,
             x = "MdaPotenciaInstaladakW", # adiciona mais uma variável
             hue = "MesReferencia")        # adiciona as outras variáveis               
```

Aumentando a sensibilidade do gráfico em relação aos dados do data frame

```python
sns.kdeplot(data = SP,
             x = "MdaPotenciaInstaladakW",
             hue = "MesReferencia", # adiciona mais uma variável
             bw_adjust = .2)        # grafico representa mais pontos dos dados 
                                    # tira a suavização do gráfico
```

Outra forma de representar o grafico de kde é por empilhamento.

```python
sns.kdeplot(data = SP,
             x = "MdaPotenciaInstaladakW",
             hue = "MesReferencia",  # adiciona mais uma variável
             bw_adjust = .2,         # tira a suavização do gráfico
             multiple = "stack")     # empilha as representações 
```

```python
sns.kdeplot(data = SP,
             x = "MdaPotenciaInstaladakW",
             hue = "MesReferencia", # adiciona mais uma variável
             bw_adjust = .2,        # tira a suavização do gráfico
             multiple = "fill")     # representatividade da densidade dos dados com relação ao geral
```

Plotagem de gráfico das retas de frequencia cumulativa

```python
sns.kdeplot(data = SP,
             x = "MdaPotenciaInstaladakW",
             hue = "MesReferencia",
             bw_adjust = .2,
             cumulative = True,    # plota gráfico de frequencia cumulativa
             common_norm = False)  # mostra individualmente todas as variáveis
```

# GRÁFICOS CATEGÓRICOS 
Os gráficos categóricos são ideias para quando  precisar mostrar uma variação de algum tipo variável, geralmente numérica com relação a uma categoria.

### CRIANDO GRÁFICO CATEGÓRICO (STRIPPLOT)
Gráfico de dispersão onde uma variável é categórica 

```python
sns.stripplot(
     data = BA.loc[lambda f: df['AnoReferencia'].isin([2006,2008,2010,2012,2014,2016,2018,2020,2022])],
     y = 'MdaPotenciaInstaladakW',
     x = 'AnoReferencia',
     hue = "MesReferencia",
     size = 10,
     marker = 'p'
 )
  
```

### CRIANDO GRÁFICO CATEGÓRICO (SWARMPLOT)
Gráfico de dispersão com os pontos não sobrepostos

```python
sns.swarmplot(
     data = BA.loc[lambda f: df['AnoReferencia'].isin([2006,2008,2010,2012,2014,2016,2018,2020,2022])],
     y = 'MdaPotenciaInstaladakW',
     x = 'AnoReferencia',
     hue = "MesReferencia",
     size = 10,
     marker = 'p',
 )
```

### CRIANDO GRÁFICO CATEGÓRICO (BOXPLOT)
Gráfico de caixa para mostrar as distribuições em relação a categoria 

```python
sns.boxplot(
     data = BA.loc[lambda f: df['AnoReferencia'].isin([2022])]
     .loc[lambda f: df['MesReferencia'].isin([3,6,9,12])],
     y = 'MdaPotenciaInstaladakW',
     x = 'AnoReferencia',
         
 )
```

### CRIANDO GRÁFICO CATEGÓRICO (VIOLINPLOT)
Gráfico de violino

```python
sns.violinplot(
     data = BA.loc[lambda f: df['AnoReferencia'].isin([2022])]
     .loc[lambda f: df['MesReferencia'].isin([3,6,9,12])],
     y = 'MdaPotenciaInstaladakW',
     x = 'AnoReferencia',
     inner = 'quartile' # Média dos valores
 )
```

```python
sns.violinplot(
     data = BA.loc[lambda f: df['AnoReferencia'].isin([2022])]
     .loc[lambda f: df['MesReferencia'].isin([3,6,9,12])],
     y = 'MdaPotenciaInstaladakW',
     x = 'AnoReferencia',
     inner = 'quartile', 
     bw = 0.2 # facilita na compreensão da estimativa da dispersão
)
```

### CRIANDO GRÁFICO CATEGÓRICO (COUNTPLOT)
Mostra as contagens de observações em cada categoria usando barras 

```python
gra1 = sns.countplot(
     data = BA,
     x = 'AnoReferencia',
     hue = 'MesReferencia',
     palette = 'BuGn_r'  
 )
gra1.figure.set_size_inches(15, 8)
```

### CRIANDO GRÁFICO CATEGÓRICO (CATPLOT)
Interface genérica para geração dos gráficos categóricos

```python
sns.catplot(
     kind = 'swarm',
     data = BA.loc[lambda f: df['AnoReferencia'].isin([2006,2008,2010,2012,2014,2016,2018,2020,2022])],
     y = 'MdaPotenciaInstaladakW',
     x = 'AnoReferencia',
     hue = "MesReferencia",
     size = 10,
     marker = 'p',
)
```

```python
sns.catplot(
     kind = 'box',
     data = BA.loc[lambda f: df['AnoReferencia'].isin([2022])]
     .loc[lambda f: df['MesReferencia'].isin([3,6,9,12])],
     y = 'MdaPotenciaInstaladakW',
     x = 'AnoReferencia',
 )
```

```python
sns.catplot(
     kind = 'violin',
     data = BA.loc[lambda f: df['AnoReferencia'].isin([2022])]
     .loc[lambda f: df['MesReferencia'].isin([3,6,9,12])],
     y = 'MdaPotenciaInstaladakW',
     x = 'AnoReferencia',
     inner = 'quartile' # Média dos valores
 )    
```
