```python
import pandas as pd
import numpy as np
```

```python
disciplinas = pd.read_excel('https://prograd.ufabc.edu.br/pdf/catalogo_disciplinas_graduacao_2021_2022_0722.xlsx')
disciplinas.head(5)
```

```python
disciplinas_B = disciplinas.query('SIGLA.str.slice(stop=1) == "B"')
disciplinas_B.head(5)
```

```python
def sorteia_disciplina(nome1, nome2, disciplinas, arquivo='lista_alunos.csv'):
  import os
  rng = np.random.default_rng()
  sorteio = disciplinas.at[disciplinas.index[rng.integers(len(disciplinas_B))],'DISCIPLINA']
  
  nova_linha = pd.DataFrame({'Nome 1':[nome1], 'Nome2':[nome2], 'Disciplina':[sorteio]})
  
  if os.path.isfile('/content/drive/MyDrive/Classroom/Bases Computacionais da Ciência 2022/'+arquivo):
    data = pd.read_csv('/content/drive/MyDrive/Classroom/Bases Computacionais da Ciência 2022/'+arquivo, index_col=False)
    data = pd.concat([data, nova_linha], axis=0)    
  else:
    data = nova_linha
  data.to_csv('/content/drive/MyDrive/Classroom/Bases Computacionais da Ciência 2022/'+arquivo, index=False)
  display(data)
  return data
```
